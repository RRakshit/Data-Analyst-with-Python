```python
import numpy as np
```

```python
# create a 100 X 20 ndarray with random integer between 0 to 5000

x = np.random.randint(5000,size=(1000,20))
print(x)
```

```python
#average value of each column of x

ave_cols = np.mean(x,axis=0)
print(ave_cols)
```

```python
# stardard deviation of each column of x

std_cols = np.std(x,axis=0)
print(std_cols)
```

```python
# print shape of average mean

ave_cols.shape
```

```python
# print shape of stardard deviation

std_cols.shape
```

```python
# mean normalization of x

X_norm =(x - ave_cols/std_cols)

# print all average from X_norm
print(X_norm)
```

```python
# print all minimum value from X_norm

print(X_norm.min(axis=0))
```

```python
# print a maximum value from X_norm
print(X_norm.max(axis=0))
```

```python
# data seperation
```

```python
# We create a random permutation of integers 0 to 4

np.random.permutation(5)
```

```python
# Create a rank 1 ndarray that contains a random permutation of the row indices of `X_norm`

row,col =X_norm.shape
row_indices =np.random.permutation(row)
print(row_indices)
```

```python
# Create a Training Set

X_train = X_norm[row_indices[0:600]]
print(X_train)
```

```python
# Create a Cross Validation Set

X_crossVal = X_norm[row_indices[600:800]]
print(X_crossVal)
```

```python
# Create a Test Set

X_test = X_norm[row_indices[800:1000]]
print(X_test)
```
