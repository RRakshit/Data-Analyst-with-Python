```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
fuel_econ=pd.read_csv('data/fuel_econ.csv')
fuel_econ.shape
fuel_econ.head()
```

```python
#create a bins for x and y axis
bin_x=np.arange(0.6,7+0.3,0.3)
bin_y=np.arange(12,58+3,3)
#create heat map or 2d histagram color minimum(cmin) for minimum value and color map(cmap) and _r for reverse
plt.hist2d(data=fuel_econ,x='displ',y='comb',cmin=0.5,cmap='viridis',bins=[bin_x,bin_y])
plt.colorbar();
```
